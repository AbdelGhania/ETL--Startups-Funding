```python
##Dependencies
import pandas as pd
import numpy as np 
from sqlalchemy import create_engine
```

```python
#Reading a CSV file "Companies_List"

companies_df = pd.read_csv("Companies_List.csv", encoding="iso-8859-1")
companies_df = companies_df.reset_index(drop=True)
```

```python
companies_df.head(5)
```

```python
# Droping some columns

companies_df = companies_df.drop(["Company Name URL", "Description", "Crunchbase Rank", "Number of Funding Rounds"], axis = 1)
```

```python
companies_df
```

```python
companies_df = companies_df.rename(columns={
    "Company Name" : "Company", "Category Groups": "Industries", "Headquarters Location":"HQ",
    "Number of Funding Rounds" : "Funding Rounds"   
})

companies_df
```

```python
# Droping another column 

companies_df = companies_df.drop("Closed Date", axis = 1)
```

```python
# adding a new coulmn 

companies_df["ID"] = list(companies_df.index)
```

```python
companies_df
```

```python
# Re-ordering columns 

companies_df = companies_df[["ID", "Company", "Industries", "HQ", "Founded Date", "Last Funding Date", "Last Funding Amount", "Last Funding Type", "Total Funding Amount", "Status"]]
```

```python
companies_df
```

```python
# Selecting companies with more than 1000000 in Total Funding
companies_df = companies_df[companies_df["Total Funding Amount"]>1000000]
companies_df
```

```python
# GroupBY Last Funding Type 
companies_df.groupby("Last Funding Type")["Total Funding Amount"].count()
```

```python
# GroupBy Companies Status

companies_df.groupby("Status")["Total Funding Amount"].count()
```

```python
# Renaming Columns

companies_df = companies_df.rename(columns={
    "ID":"id","Company" : "company", "Industries": "industries", "HQ":"hq",
    "Founded Date": "founded date","Last Funding Date": "last funding date", "Last Funding Amount": "last funding amount",
    "Last Funding Type" : "last funding type", "Total Funding Amount": "total funding amount", "Status": "status"
})

companies_df


```

```python
companies_df.to_csv("Companies_List_V1.csv")
```
